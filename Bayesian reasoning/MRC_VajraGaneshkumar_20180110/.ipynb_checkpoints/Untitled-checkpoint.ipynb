```python
data = np.loadtxt('data/train.txt')

#the last column has the values of classes, O being class failure and 1 being class success
class_data = data[:,-1]

#class_0 failure
c_f = float(np.sum(class_data==0))

#class_1 success
c_s =float(np.sum(class_data == 1))

#total number of poses
poses = 1500.00

#Probability of failure (class 0)
p_cf = c_f/poses
feature_class0 ={}
#Probability of success(class 1)
p_cs = c_s/poses
feature_class1 ={}
#*******************************Class 0*********************************

#Feature 1: Inside the table"
f1 = data[:,0]
f10_c0 = float(np.sum(f1 == 0))/c_f

f11_c0 = float(np.sum(f1 == 1))/c_f

feature_class0[1] = [f10_c0,f11_c0]

#Feature 2: 
f2 =data[:,1]
f20_c0 =float(np.sum(f2 == 0))/c_f
f21_c0 =float(np.sum(f2 == 1))/c_f
f22_c0 =float(np.sum(f2 == 2))/c_f
feature_class0[2] = [f20_c0,f21_c0,f22_c0]

#Feature 3:
f3 = data[:,2]
f30_c0 =float(np.sum(f3 == 0))/c_f
f31_c0 =float(np.sum(f3 == 1))/c_f
f32_c0 =float(np.sum(f3 == 2))/c_f
feature_class0[3] = [f30_c0,f31_c0,f32_c0]

#feature 4:
f4 = data[:,3]
f40_c0 =float(np.sum(f4 == 0))/c_f
f41_c0 =float(np.sum(f4 == 1))/c_f
f42_c0 =float(np.sum(f4 == 2))/c_f
feature_class0[4] = [f40_c0,f41_c0,f42_c0]



#********************************class 1 *********************************************
#Feature 1: Inside the table"
f1 = data[:,0]
f10_c1 = float(np.sum(f1 == 0))/c_s
f11_c1 = float(np.sum(f1 == 1))/c_s
feature_class1[1] = [f10_c1,f11_c1]

#Feature 2: 
f2 =data[:,1]
f20_c1 =float(np.sum(f2 == 0))/c_s
f21_c1 =float(np.sum(f2 == 1))/c_s
f22_c1 =float(np.sum(f2 == 2))/c_s
feature_class1[2] = [f20_c1,f21_c1,f22_c1]

#Feature 3:
f3 = data[:,2]
f30_c1 =float(np.sum(f3 == 0))/c_s
f31_c1 =float(np.sum(f3 == 1))/c_s
f32_c1 =float(np.sum(f3 == 2))/c_s
feature_class1[3] = [f30_c1,f31_c1,f32_c1]

#feature 4:
f4 = data[:,3]
f40_c1 =float(np.sum(f4 == 0))/c_s
f41_c1 =float(np.sum(f4 == 1))/c_s
f42_c1 =float(np.sum(f4 == 2))/c_s
feature_class1[4] = [f40_c1,f41_c1,f42_c1]
#*************************************************************************************
#calculating P(F) denominator(predictor prior probability)[normalizing]
x = 1
y = 1

for a,b in feature_class0.iteritems():
    for i in b:
        x = i *x
q = x * p_cf

for c,d in feature_class1.iteritems():
    for j in d:
        y = i*y
p = y*p_cs

alpha = q+p
#print alpha

#*************************************************************************************

#calculating P(F) denominator(predictor prior probability)
data_t = np.loadtxt('data/test.txt')
data_points = np.loadtxt('data/test_points.txt')

stable_points =[]
unstable_points =[]
predictions = []

for i,row in enumerate(data_t):
    sample = row[:-1]
    pro_c0 =alpha* p_cf * feature_class0[1][int(sample[0])] * feature_class0[2][int(sample[1])] \
                  * feature_class0[3][int(sample[2])] * feature_class0[4][int(sample[3])]

    pro_c1 =alpha* p_cs * feature_class1[1][int(sample[0])] * feature_class1[2][int(sample[1])] \
                  * feature_class1[3][int(sample[2])] * feature_class1[4][int(sample[3])]
        
    if pro_c1 > pro_c0 :
        predictions.append(1)
        stable_points.append()
     
   
```
