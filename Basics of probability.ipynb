1\. A box contains 100 soda bottles and 100 beer bottles. A robot needs to retreive a soda bottle from the box, but it cannot recognize the bottles while they are in the box; therefore, it has to grasp bottles from the box randomly and place them on the table before performing object recognition. Since object recognition is an expensive operation, the robot only performs this once it is fairly confident it has already placed a soda bottle on the table. How many bottles does it need to move from the box to the table before it can be at least 99.99% sure it has placed at least one soda bottle on the table?

```python
### write your code here ###
# setting up the variables with the given values
soda_bottles = 100.0
beer_bottles = 100.0
total_bottles = soda_bottles+beer_bottles
# variable to calculate the probabbility of the robot 
#picking up a beer bottle at every try
probability = soda_bottles/total_bottles
# Calculates the probability of picking up at least 
#1 soda bottle by calculating 1 - P(beer)
cur_probability = 1 - (soda_bottles/total_bottles)
# counter to keep track of the number of tries
num_tries = 1

while True:
    # picking up a beer bottle
    beer_bottles -= 1
    total_bottles -= 1
    # probability of picking up a beer bottle every time
    probability *= soda_bottles/total_bottles
    # probability of picking up at least one soda bottle every time
    cur_probability = 1 - (probability)
    print "Current probability is ", cur_probability
    # break the loop when the robot is at least 
    #99.99% sure of picking at least one soda bottle
    if cur_probability >= 0.9999:
        break
    # Incrementing the counter for every try 
    num_tries += 1
    

print "\nIt needs to take ", num_tries, "times to be at least 99.99% sure"
```

2\. A six-sided die is loaded in a way that each even face is twice as likely as each odd face. All even faces are equally likely, as are all odd faces. Construct a probabilistic model for a single roll of this die and find the probability that the outcome is less than 4.

### ANSWER 

Taking probability of getting an odd face as $x$, and probability of getting an even face is $2x$ (because each even face is twice as likely as each odd face).

As there are three odd faces\{1,3,5\} and three even faces\{2,4,6\} in a six sided dice, the total probability of all sides is given by

\begin{align*}
 P(total) &= (3 \times x) + (3 \times 2x) \\
 &= 9x 
\end{align*}

As total probability of all events is $1$, 
\begin{align*}
P(total) &= 9x = 1 \\
\implies x &= \frac{1}{9} 
\end{align*}

Now, the probability of getting less than $4$ for a single roll of the die can be represented by the events $\{1\}, \{2\}. \{3\} $
\begin{align*}
 P(1) &= P(3) = x = \frac{1}{9} \\
 P(2) &= 2x = \frac{2}{9} 
\end{align*}

\begin{align*}
P(\text{less than 4}) &= P(1) + P(2) + P(3) \\ \\
&= \frac{1}{9} + \frac{2}{9} + \frac{1}{9} \\ \\
&= \frac{4}{9}  \\
\end{align*}

For the following questions, you may have to use Bayes' rule:

\begin{equation*}
    P(A|B) = \frac{P(B|A)P(A)}{P(B)} = \frac{P(B|A)P(A)}{\sum_j P(B|A_j)P(A_j)}
\end{equation*}

where $P(A|B)$ is the conditional probability of $A$ given $B$.

For example, in the case of a binary event, we have

\begin{equation*}
    P(A|B) = \frac{P(B|A)P(A)}{P(B|A)P(A) + P(B|\neg A)P(\neg A)}
\end{equation*}

4\. A domestic robot observes the behaviour of its human everyday and uses it to adapt to the human's habits. For example, the robot has learned that there is a 92% chance that the human will drink coffee in the morning. If the human does not drink coffee in the morning, there is a 70% chance that (s)he will take a nap in the afternoon, while there is a 20% chance (s)he will take a nap if they drank coffee in the morning.

a) Which interpretation of probability is being used here and why?

b) What are the events and what are their values?

c) One morning, the robot was at its charging station and was not able to observe the human. If the robot found the human taking a nap in the afternoon, what is the probability that the human drank coffee that morning?    

### Answers 

**a) Which interpretation of probability is being used here and why?**

Conditional probability is being used here because the chances of human taking a nap in the afternoon is given based on the occurence of another event, which is, human drinking a coffee in the morning.

According to wikipedia, "conditional probability is a measure of the probability of an event given that (by assumption, presumption, assertion or evidence) another event has occurred".

**b) What are the events and what are their values?**

Let's denote the event of drinking coffee in the morning as (C), the event of not drinking coffee in the afternoon as ($\neg$C), the event of taking a nap in the afternoon as (N), and the event of not taking a nap in the afternoon as ($\neg$ N)

* Human drinking coffee in the morning

\begin{equation*}
    P(C) = 0.92
\end{equation*}

* Human not drinking coffee in the morning

\begin{equation*}
    P(\neg C) = 0.08
\end{equation*}

* Human taking a nap in the afternoon given the human does not drink coffee in the morning

\begin{equation*}
    P(N |  \neg C) = 0.70
\end{equation*}

* Human taking a nap in the afternoon given the human drank coffee in the morning

\begin{equation*}
    P(N | C) = 0.20
\end{equation*}

The events and their values are:

* Human drinking coffee in the morning and taking a nap in the afternoon

\begin{align*}
    P(C \land N) &= P(N|C) \cdot P(C) \\
    &= 0.2 \times 0.92 \\
    &= 0.184
\end{align*}

* Human drinking coffee in the morning and not taking a nap in the afternoon

\begin{align*}
    P(C \land \neg N) &= P(\neg N|C) \cdot P(C) \\
    &= 0.8 \times 0.92 \\
    &= 0.736
\end{align*}

* Human not drinking coffee in the morning and taking a nap in the afternoon

\begin{align*}
    P(\neg C \land N) &= P(N|\neg C) \cdot P(\neg C) \\
    &= 0.7 \times 0.08 \\
    &= 0.056
\end{align*}

* Human dnot rinking coffee in the morning and not taking a nap in the afternoon

\begin{align*}
    P(\neg C \land \neg N) &= P(\neg N| \neg C) \cdot P(\neg C) \\
    &= 0.3 \times 0.08 \\
    &= 0.024
\end{align*}


**c) One morning, the robot was at its charging station and was not able to observe the human. If the robot found the human taking a nap in the afternoon, what is the probability that the human drank coffee that morning?**

Let's denote the event of drinking coffee in the morning as (C), the event of not drinking coffee in the afternoon as ($\neg$C), the event of taking a nap in the afternoon as (N), and the event of not taking a nap in the afternoon as ($\neg$ N) 

Applying Bayes' rule,

\begin{align*}
 P(\text{coffee in the morning | nap in the afternoon}) &= \frac{P(N|C)P(C)}{P(N)} \\ \\
 &= \frac{P(N|C)P(C)}{\sum_i P(N|C_i)P(C_i)} \\ \\
 &= \frac{P(N|C)P(C)}{P(N|C)P(C) + P(N|\neg C)P(\neg C)} \\ \\
 \text{Plugging in the values,} \\ \\
 &= \frac{0.2 \times 0.92}{(0.2 \times 0.92)+(0.7 \times 0.08)} \\ \\
 &= 0.7667 \\ \\
\end{align*}
